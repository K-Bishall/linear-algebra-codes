# Pearson Correlation Coefficient
### Using movies dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Read the data, and visualize

```python
data = pd.read_csv('movies-data.csv')
data.head(6)
```

```python
gross = data["gross"].to_numpy()
budget = data["budget"].to_numpy()
votes = data["votes"].to_numpy()

print(gross[:5])
```

```python
plt.scatter(budget, gross, alpha=0.5, color='blue')
plt.xlabel("Budget")
plt.ylabel("Gross Earnings")
plt.title("Budget vs Gross Earnings")
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10, 10))
plt.scatter(votes, gross)
```

### Pearson Correlation Coefficient Calculation

```python
def correlation_coeff(x: np.ndarray, y: np.ndarray) -> float:
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    x_vector = x - x_mean
    y_vector = y - y_mean

    dot = np.dot(x_vector, y_vector)
    norm_x = np.linalg.norm(x_vector)
    norm_y = np.linalg.norm(y_vector)

    return dot / (norm_x * norm_y)
```

```python
budget_gross_coeff = correlation_coeff(budget, gross)
print(budget_gross_coeff)
```
