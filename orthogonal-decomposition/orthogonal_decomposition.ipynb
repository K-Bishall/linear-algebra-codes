# Orthogonal Vector Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define ref and target vectors

```python
# define target and ref vectors
target_vec = np.random.randint(1, 10, 2)
ref_vec = np.random.randint(1, 10, 2)
```

### Implement orthogonal vector decomposition

```python
# compute parallel and perpendicular components
parallel_vec = ref_vec * (np.dot(target_vec, ref_vec) / np.dot(ref_vec, ref_vec))
perp_vec = target_vec - parallel_vec
```

### Plot all the vectors

```python
# plot the vectors
plt.figure(figsize=(6, 6))
plt.grid()

origin = np.array([0, 0])
arrow_options = dict(width=0.06, head_width=0.2)

plt.arrow(*origin, *target_vec, color='r', **arrow_options)
plt.arrow(*origin, *ref_vec, color='g', **arrow_options)

target_label_pos = target_vec + np.array([0.1, 0.1])
ref_label_pos = ref_vec + np.array([0.1, 0.1])
plt.text(*target_label_pos, s=r'$\vec{T}$')
plt.text(*ref_label_pos, s=r'$\vec{R}$')

plt.arrow(*origin, *parallel_vec, color='b', **arrow_options)
plt.arrow(*origin, *perp_vec, color='y', **arrow_options)

parallel_label_pos = parallel_vec + np.array([0.1, 0.1])
perp_label_pos = perp_vec + np.array([0.1, 0.1])
plt.text(*parallel_label_pos, s=r'$\vec{T} \parallel \vec{R}$')
plt.text(*perp_label_pos, s=r'$\vec{T} \perp \vec{R}$')

plt.axis('equal')
plt.show()
```
