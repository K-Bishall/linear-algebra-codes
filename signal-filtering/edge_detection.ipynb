# Edge Detection of Time Series Signal

- Define a plateau signal, and kernel of certain length
- Plot and visualize the kernel and the signal
- Loop over the points in signal
- At each point, compute dot product between kernel and time series data having the same length as kernel
- Plot and visualize the result
- Play around by changing kernel

![Steps](edge-detection-steps.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define signal, kernal, and visualize

```python
# kernel for edge detection
kernel = np.array([-1, 1])

# and the "signal" (a plateau)
signal = np.zeros(30)
signal[10:20] = 1
```

```python
# plot them
_, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(kernel, 'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-5, 5])

axs[1].plot(signal, 'ks-')
axs[1].set_title('Time series signal')

plt.show()
```

### Dot product of kernel and signal

```python
# initialize the result map as zeros
edge_detection_result = np.zeros(len(signal))

# loop over the signal and do template-matching (via dot products!)
for t in range(1, len(signal) - 1):
    edge_detection_result[t] = np.dot(kernel, signal[t - 1:t + 1])
```

### Visualize the result

```python
# plot the result
plt.figure(figsize=(6, 4))

plt.plot(signal, 'ks-', label='Signal', linewidth=3)
plt.stem(range(len(edge_detection_result)), edge_detection_result, markerfmt='o', basefmt=' ', linefmt='',
         label='Edge Detection')
plt.legend()

plt.show()
```
