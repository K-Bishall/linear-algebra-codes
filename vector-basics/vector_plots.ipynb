# Drawing Vectors

```python
import numpy as np
import matplotlib.pyplot as plt

# some plot configs
plt.rcParams['font.size'] = 14
plt.rcParams['axes.linewidth'] = 2
```

```python
# define vectors
vec_a = np.array([5, 6])
vec_b = np.array([1, 5])
```

```python
plt.figure(figsize=(6, 6))
plt.axis('equal')
plt.grid()
origin = np.array([0, 0])

arrow_options = dict(width=0.06, head_width=0.3, length_includes_head=True)

# draw vectors
plt.arrow(*origin, *vec_a, color='red', **arrow_options)
plt.arrow(*origin, *vec_b, color='b', **arrow_options)

# adding texts
label_a_pos = vec_a + np.array([0.1, 0.1])
label_b_pos = vec_b + np.array([0.1, 0.1])
plt.text(*label_a_pos, s=r'$\vec{A}$', fontsize=12)
plt.text(*label_b_pos, s=r'$\vec{B}$', fontsize=12)

vec_c = vec_a + vec_b
label_c_pos = vec_c + np.array([0.1, 0.1])
plt.arrow(*origin, *vec_c, color='g', **arrow_options)
plt.text(*label_c_pos, s=r'$\vec{B}$', fontsize=12)

plt.show()
```
